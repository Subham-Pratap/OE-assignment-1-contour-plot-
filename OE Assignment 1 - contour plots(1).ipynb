```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-5,5,100)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, (0-5*x)/4, '-r', label='y=-5x/4')
plt.plot(x, (4-5*x)/4,'-.g', label='y=4-5*x/4')
plt.plot(x, (13-5*x)/4,':b', label='y=13-5*x/4')
plt.plot(x, (18-5*x)/4,'--m', label='y=18-5*x/4')
plt.plot(x, (20-5*x)/4,'--b', label='y=20-5*x/4')
plt.plot(x, (21-5*x)/4,':y', label='y=21-5*x/4')
plt.plot(x, (25-5*x)/4,'-b', label='y=25-5*x/4')
plt.legend(loc='upper left')
plt.show()

```
